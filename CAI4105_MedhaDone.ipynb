```python
import numpy as np
import matplotlib.pyplot as plt

# Number of observations
N = 10

x = np.array([
    0.0, 0.11111111, 0.22222222, 0.33333333, 0.44444444,
    0.55555556, 0.66666667, 0.77777778, 0.88888889, 1.0
])

# True function values
true_values = np.sin(2 * np.pi * x)

# Add Gaussian noise with mean 0 and variance 0.2 (std = sqrt(0.2))
noise = np.random.normal(0, np.sqrt(0.2), N)

# Targets
t = np.array([
    -0.6822356, 0.33335691, 0.96442107, 0.97485011, 0.2341361,
    -0.18457618, -1.42573085, -0.33913659, -0.67952673, 0.49966989
])



# visualize
plt.scatter(x, t, label="Noisy observations")
plt.plot(x, true_values, label="True function (sin)", color="red")
plt.legend()
plt.show()
```

```python

# Polynomial function
def f(x, w):
    """
    Evaluate f(x, w) = sum_{j=0}^M w_j x^j for vector x.
    """
    x = np.asarray(x)
    powers = np.vstack([x**j for j in range(len(w))]).T
    return powers @ w
```

image.png

```python
M = 3  # choose the polynomial order
w_star = np.array([-0.72345793, 15.03173393, -40.50824376,  26.74805169])


# Training error E
y_pred = f(x, w_star)
E = 0.5 * np.mean((y_pred - t) ** 2)

#print error
print('Training error is:', E)
print('The training error rounded is approximately', round(E, 2))

# Visualization against the true function
xx = np.linspace(0, 1, 10)
plt.scatter(x, t, label="Observations")
plt.plot(xx, np.sin(2*np.pi*xx), label="True function sin(2πx)")
plt.plot(xx, f(xx, w_star), label=f"Polynomial fit (M={M})")
plt.legend()
plt.xlabel("x")
plt.ylabel("t / f(x)")
plt.title("Polynomial Least Squares Fit")
plt.show()
```
