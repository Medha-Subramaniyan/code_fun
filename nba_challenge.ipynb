# PER
- normalized stat to compare different players across different teams 
- summarizes a player's overall performance by considering both positive and negative contributions to a game
- per-minute statistic
  - adjusted for how much time a player is on the court
  - also pace-adjusted to account for different team tempo


# USAGE RATE
- estimates the percentage of a team's possessions a player uses while they are on the court
-  quantifies a player's contribution to a team's offensive possessions
- OFFENSIVE

# TS% 
- player's shooting efficiency that considers all types of shots (2-pointers, 3-pointers, and free throws)
- more comprehensive than FG% becuase it takes 3pts in account

```python
from nba_api.stats.endpoints import LeagueDashPlayerStats

#1. build params 
params = { 
    "season": "2016-17",
    "season_type_all_star": "Regular Season",
    "per_mode_detailed": "PerGame",
    "plus_minus": "N",
    "pace_adjust": "N",
    "rank": "N",
}

#2. get the data (pass params ass arguent and use get_data_frames())
endpoint = LeagueDashPlayerStats(**params)
df_list  = endpoint.get_data_frames()
player_stats_df = df_list[0]
```

```python
#3. turn into a df
stats = pd.DataFrame(player_stats_df)
stats# final dataframe
```

## Descriptive Analytics
(“What happened?” / Summary & reporting)
These features help you summarize and track performance over time or across groups.

# Normalized rates
PTS_per36, AST_per36, REB_per36: raw counts scaled to 36 minutes.

# Efficiency splits
TS% (True Shooting%), eFG%, plus separate 2P% and 3P% for deeper shooting breakdowns.

# Rolling aggregates
rolling_PTS_avg_5g: 5‑game moving average of points.
rolling_TS%_10g: last‑10‑game shooting efficiency.

# Cumulative totals
cum_PTS_season: running total of points through each date.

# Situational flags
is_clutch: game within 5 points in final 5 minutes (0/1).
home_game vs. away_game binary.

# Simple ratios
AST_to_TOV: assists divided by turnovers.
ORB_rate or DRB_rate: percentage of available rebounds grabbed.

# Classification / Regression
(“What will happen?” / Predict outcomes or values)
Here, you build features that carry predictive signal for a target (win/loss, next‑game PTS, salary band, etc.).

# Usage & possession metrics
Usage_Rate: share of team possessions used by the player.
 ORtg / DRtg: individual offensive/defensive ratings.

# Advanced composites
PER (Player Efficiency Rating), RAPTOR, VORP — league‑standard impact estimates.

# Interaction terms
 MP × TS%: captures volume × efficiency synergy.
3PA × 3P%: high‑volume shooters weighted by accuracy.
# Lagged and trend features
PTS_last_game, AST_rolling_diff: difference between last two rolling averages.
game_to_game_std_PTS: volatility in scoring—predict consistency outcomes.

# Contextual encodings
One‑hot position (PG/SG/SF/PF/C).
opp_def_rank_bin: opponent defense rating bucketed (top‑5, middle, bottom‑5).
# Team‑level joins
team_pace, team_off_rating joined back to individual lines so model knows system context.

# 3. Clustering / Profiling
(“How do entities group?” / Player archetypes & segmentation)
Features here feed unsupervised algorithms (k‑means, hierarchical clustering, UMAP) to reveal natural groupings.

# Profile vectors
[Usage_Rate, TS%, AST_rate, REB_rate, STL_rate, BLK_rate] normalized per‑36 or per‑100 possessions.

# Shooting‑mix percentages
pct_rim_shots, pct_midrange, pct_3pt_attempts: the share of each shot zone.

# Playstyle ratios
ast_to_usage, blk_to_usage, stl_to_usage: defensive vs. offensive tendencies.

# Dimensionality reduction inputs
PCA on the above rates to get “style components” (e.g. Component 1 = paint‑heavy vs. perimeter).

# Role flags
starter_flag (MP > 25 minutes) vs. bench_flag.
clutch_specialist if clutch FG% > season average + threshold.

# Synergy / lineup metrics
net_rating_when_on_court, teammate_pass_share: how a player fits with others.

# Picking & testing
## Match features to your goal:

Descriptive needs simple aggregates & rates.

Predictive tasks thrive on interaction, context flags, and lagged signals.

Clustering leans on normalized style vectors and dimensionality‑reduced representations.

## Quickly prototype:
 run a tiny regression or clustering on a subset to see which features carry signal.

## Iterate & refine
 drop the noisy ones, enrich the strong ones, and always ask “does this help me answer my core question?”
